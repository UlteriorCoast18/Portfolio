# 01 - EDA Payroll / HR Dataset
This notebook performs initial exploratory data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = '../data/raw/dataset.csv'  # update your dataset name
hr = pd.read_csv(file_path)
hr.head()
```

## Basic Info

```python
hr.info()
```

## Descriptive Statistics

```python
hr.describe(include='all')
```

## Missing Values

```python
hr.isnull().sum().sort_values(ascending=False)
```

## Distribution Plots

```python
numeric_cols = hr.select_dtypes(include=np.number).columns
hr[numeric_cols].hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(hr.corr(), annot=False, cmap='coolwarm')
plt.show()
```
