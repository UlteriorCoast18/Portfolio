# Introduction

This is the **Workforce Intelligence & Payroll Optimization System** project. In this introduction we will first talk a little bit about the origin of the dataset and some changes that were necessary to made it in order for the personally identifiable information not to be used.

The original raw dataset contained personally identifiable information (PII), including:

- First Name
- Last Name
- Middle Initial

```python
import pandas as pd

df = pd.read_csv('../data/raw/payroll_raw.csv')

cols_to_remove = [
    'First Name',
    'Last Name',
    'Middle Init'
]

df[cols_to_remove].head()
```

These fields were removed intentionally as part of the data cleaning process to protect individual privacy and to ensure the dataset can be safely published and used for analysis.

Only anonymized fields, such as Employee Identifier, were preserved.
The cleaned dataset is exported to `data/processed/payroll_clean.csv`.

```python
df_clean = df.drop(columns=cols_to_remove)
df_clean.to_csv('../data/processed/payroll_clean.csv', index=False)
```

All personally identifiable information (PII) such as employee names was removed.
The dataset is now safe for public use and analysis.
